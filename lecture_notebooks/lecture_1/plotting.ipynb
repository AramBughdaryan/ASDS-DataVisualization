```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
np.random.seed(42)
x_uniform = np.random.random(250)
```

```python
x_uniform.shape
```

```python
plt.hist(x_uniform)
plt.show()
```

```python
plt.figure(figsize = (10, 6))
plt.hist(x_uniform, bins=90)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.hist(x_uniform, bins=30, color='skyblue', edgecolor='red')
plt.title('Histogram', fontsize=16)
plt.xlabel('Value', fontsize=12,)
plt.ylabel('Frequency', fontsize=12)
plt.grid(True)
plt.show()
```

```python
fig = plt.figure(figsize = (10, 6))
ax = fig.add_subplot(1, 1, 1) # nrows, ncol and index
ax.set_title('Histogram')
ax.set_xlabel('Value')
ax.set_ylabel('Frequency')
ax.hist(x_uniform, color='green')
plt.show()
```

```python
sns.displot(x_uniform, kind="kde"); # kernel density estimation
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(x_uniform, kde=True, bins=30, color='purple')
plt.title('Distplot', fontsize=16)
plt.xlabel('Value', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(True)
plt.show()
```

```python
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(1, 1, 1)
ax.scatter(a, b, s=d)
plt.show()
```

```python
np.random.seed(42)
x1 = np.random.randint(20, 40, 5)
x2 = np.random.randint(15, 35, 5)
x3 = np.random.randint(5, 15, 5)
```

```python
total = x1 + x2 + x3
x1_perc = 100*(x1/total)
x2_perc = 100*(x2/total)
x3_perc = 100*(x3/total)
```

```python
x1_perc
```

```python
fig = plt.figure(figsize=(16, 10))
ax = fig.add_subplot(1, 1, 1)

names = ['A', 'B', 'C', 'D', 'E']
ind = np.arange(5)
plt.xticks(ind, names)

ax.bar(ind, x1_perc)
ax.bar(ind, x2_perc, bottom = x1_perc)
ax.bar(ind, x3_perc, bottom = x1_perc + x2_perc)

plt.show()
```

## line plot

```python
np.random.seed(42)
x = np.random.normal(1, 1, 100) + np.linspace(1, 15, 100)
y = np.random.normal(1, 1, 100) + np.linspace(10, 20, 100)
```

```python
fig = plt.figure(figsize=(16, 10))
ax = fig.add_subplot(1, 1, 1)

ax.plot(x)
plt.show()
```

```python
fig = plt.figure(figsize=(16, 10))
ax = fig.add_subplot(1, 1, 1)

ax.fill_between(np.linspace(1, 15, 100), x, color="red")
ax.fill_between(np.linspace(1, 15, 100), y, color = "blue",alpha=0.8)
# ax.fill_between(np.linspace(1, 15, 100), x, color="red",  alpha=0.5)


plt.show()
```

```python
y1=np.concatenate([y[:50],y[50:]-15])
```

```python
fig = plt.figure(figsize=(8, 5))
ax = fig.add_subplot(1, 1, 1)

ax.fill_between(np.linspace(1, 15, 100), x, y1, color="red")


plt.show()
```

# subplots

```python
np.random.seed(42)
x1 = np.random.exponential(10, 92) + np.linspace(1, 20, 92)
x2 = np.random.uniform(1, 25, 92) - np.linspace(1, 30, 92)
x3 = np.random.normal(1, 1, 92) - np.linspace(1, 15, 92)
x4 = np.random.normal(1, 2, 92)
```

```python
fig = plt.figure(figsize=(8, 5))
ax = fig.add_subplot(1, 1, 1) # nrow, ncol, index

ax.plot(x1)
ax.plot(x2)
ax.plot(x3)
ax.plot(x4)
plt.show()
```

```python
fig = plt.figure(figsize=(16, 10))

ax = fig.add_subplot(2, 2, 1) # ncol, nrow,idex
ax.plot(x1)

ax = fig.add_subplot(2, 2, 2)
ax.plot(x2)

ax = fig.add_subplot(2, 2, 3)
ax.plot(x3)

ax = fig.add_subplot(2, 2, 4)
ax.plot(x4)

plt.show()
```

```python
fig = plt.figure(figsize=(16, 10))
ax1 = fig.add_subplot(2, 2, 1) # ncol, nrow,idex
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
ax4 = fig.add_subplot(2, 2, 4)

ax1.plot(x1)
ax2.plot(x2)
ax3.plot(x3)
ax4.plot(x4)
plt.show()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(8, 5))

ax11 = axes[0][0]
ax12 = axes[0][1]
ax21 = axes[1][0]
ax22 = axes[1][1]

ax11.plot(x1, c='blue')
ax12.plot(x2, c='orange')
ax21.plot(x3, c='green')
ax22.plot(x4, c='brown')

ax11.set_ylabel('Sales')
ax21.set_ylabel('Sales')

ax21.set_xlabel('Time')
ax22.set_xlabel('Time')

plt.show()
```

```python
t = np.linspace(0, 10, 100)

fig = plt.figure(figsize=(8,5))
ax1 = fig.add_subplot(1, 1, 1)
# ax2 = ax1.twinx()

ax1.plot(t, t**2, 'b')
ax1.plot(t, 100/(t + 0.1), 'r')
ax1.set_ylabel('Value 1', color='red')
# ax2.set_ylabel('Value 2', color='blue')
ax1.set_xlabel('Time');
```

```python
t = np.linspace(0, 10, 100)

fig = plt.figure(figsize=(16,10))
ax1 = fig.add_subplot(1, 1, 1)
ax2 = ax1.twinx()

ax1.plot(t, t**2, 'b')
ax2.plot(t, 100/(t + 0.1), 'r')
ax1.set_ylabel('Value 1', color='red')
ax2.set_ylabel('Value 2', color='blue')
ax1.set_xlabel('Time');
```
